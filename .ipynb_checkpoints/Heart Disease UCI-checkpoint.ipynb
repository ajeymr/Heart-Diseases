# Heart Disease UCI

### This project looks to do an exploratory analysis of a database of patients who are with heart diseases. The data for this analysis is taken from the website https://archive.ics.uci.edu/ml/datasets/Heart+Disease

```python
# importing libraries

import pandas as pd
import numpy as np
```

```python
# importing a .csv file from the source folder

filename = "cleveland processed original.csv"

df = pd.read_csv(filename)
print (" Data imported ")
```

```python
# displaying a part of the table 
print (df.shape)
df.head(5)
```

```python
# Counting the number of missing value

df.isin(['?']).sum(axis=0)
```

```python
# replace "?" to NaN

df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
# Checking oif the question marks have gone

df.isin(['?']).sum(axis=0)
```

```python
# counting the nan values

df.isna().sum()
```

```python
# dropping the rows with nan

df.dropna(inplace = True)
```

```python
# counting the nan values.

df.isna().sum()
```

```python
# shape after the 

print (df.shape)
df.head(10)
```

```python
#checking the datatypes

df.dtypes
```

```python
# changing the datatypes from object to int

df[["ca","thal"]] = df[["ca","thal"]].astype("int")
```

```python
#rechecking the datatypes

df.dtypes
```

```python
# importing libraries

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
#heat map of the correlation data

# ax = plt.subplots(figsize=(10,7))
# sns.heatmap(df.corr(),cmap="YlGnBu",annot=True)

plt.rcParams['figure.figsize'] = (20, 15)
plt.style.use('ggplot')

sns.heatmap(df.corr(), annot = True, cmap = 'Wistia')
plt.title('Heatmap for the Dataset', fontsize = 20)
```

```python
# regression plot with age and cholestrol

age_chol = df[['age','chol']]
age_chol.head()

sns.regplot(x='age', y='chol', data=age_chol)
```

```python
# creating a table with a count of the different people with different levels of heart disease severity

target = df.groupby('target', axis=0).count()
target = target.drop(['sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
       'exang', 'oldpeak', 'slope', 'ca', 'thal'], axis = 1) 
target.rename(columns={'age':'Number of people'}, inplace=True)
target.head()

# creating a pie chart

#colors for the 
colors_list = ['gold', 'yellowgreen', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0, 0, 0.1, 0.1, 0.1] # ratio for each continent with which to offset each wedge.

target['Number of people'].plot(kind='pie',
                            figsize=(15, 6),
                            autopct='%1.1f%%', 
                            startangle=90,    
                            shadow=True,       
                            labels=None,         # turn off labels on pie chart
                            pctdistance=1.12,    # the ratio between the center of each pie slice and the start of the text generated by autopct 
                            colors=colors_list,  # add custom colors
                            explode=explode_list # 'explode' lowest 3 continents
                            )

# scale the title up by 12% to match pctdistance
plt.title('People with heart conditions', y=1.12) 

plt.axis('equal') 

# add legend
plt.legend(labels=target.index, loc='upper left') 

plt.show()
```

```python
# 0 = no heart condition
# 1 = heart condition
# changing the targets 1,2,3,4 to 1 

df['target'][df['target'] == 1] = 1
df['target'][df['target'] == 2] = 1
df['target'][df['target'] == 3] = 1
df['target'][df['target'] == 4] = 1

#verifying the above operation

df['target'].value_counts()
```

```python
# Finding the average parameters for people with heart condition and people without

df.groupby('target').mean()
```

```python
# creating a new database for only with the heart condition patients

disease = df[df['target'] == 1]
disease.drop(['target'], axis = 1, inplace = True)
disease.head()
```

```python
# distribution of age for people with heart condition

plt.style.use('fivethirtyeight')
plt.rcParams['figure.figsize'] = (15, 5)
sns.distplot(disease['age'], color = 'cyan')
plt.title('Distribution of Age for Heart condition patients', fontsize = 20)
plt.show()
```

```python
# getting libraries
from sklearn import model_selection
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
# creating x and y dataframes for applying the ML algorithm
x = df[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
       'exang', 'oldpeak', 'slope', 'ca', 'thal']]
y = df[['target']]

x_shape = x.shape
y_shape = y.shape
print("The x has Rows {}, Columns {}".format(x_shape[0], x_shape[1]))
print("The y has Rows {}, Columns {}".format(y_shape[0], y_shape[1]))
```

```python
# splitting the test train
x_trainset,x_testset,y_trainset,y_testset = train_test_split(x, y, test_size = 0.20, random_state = 3)
print("The x_trainset has Rows {}, Columns {}".format(x_trainset.shape[0], x_trainset.shape[1]))
print("The y_trainset has Rows {}, Columns {}".format(y_trainset.shape[0], y_trainset.shape[1]))
print("The x_testset has Rows {}, Columns {}".format(x_testset.shape[0], x_testset.shape[1]))
print("The y_testset has Rows {}, Columns {}".format(y_testset.shape[0], y_testset.shape[1]))
```

```python
heartdisease = DecisionTreeClassifier(criterion="entropy")
heartdisease.fit(x_trainset,y_trainset)
```
